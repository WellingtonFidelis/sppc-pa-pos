```python
from faker import Faker
from datetime import datetime, timedelta
from random import randint
import requests
```

```python
fake = Faker()
```

```python
def gerar_consumidor():
    dt_ini = fake.date_between(start_date=datetime(year=2023, month=1, day=1), end_date=datetime(year=2024, month=12, day=31))
    dt_end = dt_ini + timedelta(randint(50, 75))
    weigth_ini = randint(50, 90)
    weigth_end = weigth_ini + randint(1, 10) if weigth_ini < 60 else weigth_ini - randint(1, 10)
    return {
        "descricao": fake.sentence(nb_words=2).replace(".", ""),
        "data_inicio": dt_ini,
        "data_fim": dt_end,
        "peso_inicial": weigth_ini,
        "peso_final": weigth_end,
        "eficaz": fake.boolean(),
        "consumidores": randint(5, 55),
        "produtos": [randint(3, 19) for _ in range(randint(1, 3))],
    }
```

```python
tratamentos = [ gerar_consumidor() for _ in range(2000)]
```
